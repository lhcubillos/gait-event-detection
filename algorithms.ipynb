```julia
subjects = ["AB06", "AB07", "AB08"]
activities = ["treadmill", "levelground", "ramp","stair"]
d_mistakes = Dict()
d_rmse = Dict()
d_rmse_baseline = Dict()
d_avg_time_dist = Dict()
for subject in subjects
    for activity in activities
        for (trial_num,trial) in enumerate(dict[subject][activity])
            printstyled("$subject\t$activity\t$trial_num\n", color=:blue)
            t = trial["imu"][:,1]
            ground_truth_hs = trial["gcRight"][:,2]
            imu_shank = trial["imu"][:,12]
            heel_strike_phase, mistakes, rmse, rmse_baseline, avg_time_dist  = getPhaseGreene(t, imu_shank, ground_truth_hs, 200)
            d_mistakes[(subject, activity, trial_num)] = mistakes
            d_rmse[(subject, activity, trial_num)] = rmse
            d_rmse_baseline[(subject, activity, trial_num)] = rmse_baseline
            d_avg_time_dist[(subject, activity, trial_num)] = avg_time_dist
        end
    end
end
# plot(t,heel_strike_phase, label="Estimated")
# plot!(t, ground_truth_hs, label="Original")
```

```julia
function get_average(dict; activity=nothing, subject=nothing)
    filter_dict = dict
    if !isnothing(activity)
        filter_dict = filter(p -> p.first[2] == activity, filter_dict)
    end
    if !isnothing(subject)
        filter_dict = filter(p -> p.first[1] == subject, filter_dict)
    end
    return mean(values(filter_dict))
end
for activity in activities
    rmse = get_average(d_rmse, activity=activity)
    rmse_baseline = get_average(d_rmse_baseline, activity=activity)
    println("$activity -- rmse: $rmse, rmse_baseline: $rmse_baseline")
end
```
